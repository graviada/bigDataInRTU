```python
btc_data = pd.read_csv('bitcoin.csv')
btc_data.head()
```

```python
projection = 14
btc_data['predict'] = btc_data['close'].shift(-projection)
btc_data
```

```python
X = btc_data['close']
y = btc_data['predict']
X = np.array(X, type(float))
y = np.array(y, type(float))
plt.figure(figsize=(10, 10))
plt.grid(True)
plt.title('Диаграмма рассеяния', fontsize = 20)
plt.xlabel('Predict')
plt.ylabel('Close')
plt.scatter(X, y, alpha = 0.5)
plt.show()
```

```python
X = pd.DataFrame(btc_data.tail(14), columns=['volumefrom'])
Y = pd.DataFrame(btc_data.tail(14), columns=['volumeto'])
X = np.array(X, type(float))
Y = np.array(Y, type(float))
```

```python

plt.grid(True)
plt.scatter(X, Y, color='g', alpha=0.3)
plt.title("Предсказания стоимости криптовалюты")
plt.xlabel("volumefrom")
plt.ylabel("volumeto")
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, Y)
print("Наклон линии регрессии: ", regression.coef_)
print("y-перехват: ", regression.intercept_)
```

```python
plt.grid(True)
plt.scatter(X, Y, color='purple', alpha=0.3)
plt.plot(X, regression.predict(X), color='yellow')
plt.title("Предсказания стоимости криптовалюты")
plt.xlabel("volumefrom")
plt.ylabel("volumeto")
plt.show()
```

```python
DF = pd.read_csv('housePrice.csv', delimiter=',')
DF['Address'].fillna('other', inplace=True)
print(DF.head())
```

```python
for column in DF.columns:
    missing = np.mean(DF[column].isna() * 100)
    print(f" {column} : {round(missing, 1)}%")
```

```python
DF["Area"] = pd.to_numeric(DF["Area"], errors='coerce')
DF["Price(USD)"] = pd.to_numeric(DF["Price(USD)"], errors='coerce')
DF["Area"].fillna(DF["Area"].median(), inplace=True)
```

```python
X = pd.DataFrame(DF, columns=['Area'])
Y = pd.DataFrame(DF, columns=['Price(USD)'])
X = np.array(X, type(float))
Y = np.array(Y, type(float))
```

```python
print("Коэффициент корреляции: ", DF['Area'].corr(DF['Price(USD)']))
n = np.size(X)
```

```python
SSxy = np.sum(Y * X) - n * np.mean(Y) * np.mean(X)
SSxx = np.sum(X * X) - n * np.mean(X) * np.mean(X)

B1 = SSxy/SSxx
B0 = np.mean(Y) - B1 * np.mean(X)

print("Коэффициенты: ", "B1 = ", B1, "B0 = ", B0)
```

```python
plt.scatter(X, Y, color='m', marker='o', s=30)
y_pred = B0 + B1 * X
```

```python
regression = LinearRegression()
regression.fit(X, Y)
print("Наклон линии регрессии: ", regression.coef_)
print("y-перехват: ", regression.intercept_)
```

```python
plt.grid(True)
plt.plot(X, y_pred, color='g')
plt.title("Зависимость между стоимостью и площадью квартир")
plt.xlabel("Area")
plt.ylabel("Price(USD)")
plt.show()
```
